# Imagens são feitas do quê?

Se você pegar qualquer imagem, independente da resolução, e aplicar um zoom de aproximação por tempo o bastante, você vai começar a ver o que chamamos de _efeito tabuleiro de xadrez_. Estes pequenos quadrados, quando não mais podem ser divididos em outros quadrados menores, são chamados de _pixels_. Essa palavra se originou da junção das palavras ***pic**ture **el**ement*, o que nos permite compreender o que de fato é um pixel; O menor elemento de uma imagem. Se você está interessado em curiosidades sobre o pixel como "Por que os pixels são quadrados?", "Como cabem mais pixels em espaços iguais ou menores?" ou "Por que os pixels nas câmeras de smartphones têm tamanhos diferentes?" clique <a href="http://www.tecmundo.com.br/pixel/45102-por-que-os-pixels-sao-quadrados-e-outras-curiosidades.htm" target="_blank">aqui</a>. Informações mais técnicas sobre os pixels podem ser encontradas também clicando <a href="http://www.tecmundo.com.br/pixel/7529-pixel-o-que-voce-precisa-saber-sobre-ele-.htm" target="_blank">aqui</a>. Abaixo, podemos ver uma imagem "pixelizada":
<img width=300px src="http://www.publicdomainpictures.net/pictures/130000/velka/pixel-couple.jpg"><center><h6>Fonte: http://www.publicdomainpictures.net/</h6></center>

<br>Uma das razões para esta imagem estar _pixelizada_ pode ter sido por uma questão de quantização espacial, isto é, em vez de uma imagem com resolução de 600x400 (240.000 pixels), a matriz de pixels continha apenas 2.400 pixels, algo como uma resolução de 60x40. Quanto maior a _resolução espacial_, maior a riqueza de detalhes da imagem. No entanto, quanto maior a resolução espacial, mais espaço será necessário para armezenar a imagem. Peguemos um exemplo de uma imagem <a href="https://pt.wikipedia.org/wiki/N%C3%ADvel_de_cinza" target="_blank">monocromática</a> com até 256 tons de cinza representados, isto é, cada pixel terá 8 bits (2^8 = 256 tons), o que equivale a 1 byte. Se essa imagem tiver uma resolução de 640x480, nós teremos 307.200 pixels, ou 307.200 bytes. Para saber o espaço em kilobytes necessário para armazenar esta imagem, dividiremos o valor em bytes por 1024 e concluiremos que precisamos de 300K ou 300 kilobytes para armazenar essa imagem. Se a imagem tivesse uma maior resolução espacial, como 1024x1024, precisaríamos de 1M ou 1 megabyte para armazenar a imagem, pois 1024x1024 = 1.048.576 bytes, que dividido por 1024 nos traz 1024K ou 1M. A quantidade de bits por pixel (bpp) é também chamada de _color depth_.

Costumamos dizer que uma imagem vale mais que mil palavras. Com certeza, dependendo da <a href="http://mepixa.com.br/blog/uma-breve-aula-sobre-resolucao-de-imagem" target="_blank">resolução da imagem</a>, ela pode valer mais que mil pixels :-). E dependendo do <a href="http://www.tecmundo.com.br/video/2481-o-que-e-espaco-de-cores-.htm" target="_blank">espaço de cores</a> utilizado, isto é, quantos bits cada pixel ocupa na memória do computador, pode sim valer mais que mil <a href="https://pt.wikipedia.org/wiki/Palavra_(ci%C3%AAncia_da_computa%C3%A7%C3%A3o)" target="_blank">palavras</a>. # Matriz de Pixels

Em computacão, frequentemente ocorre de acharmos que as estruturas são como as percebemos visualmente. Uma estrutura de dados dinâmica estando alocada de forma contígua na memória, ou matrizes algébricas armazenadas em suas respectivas dimensões. No entanto, não é isso que costuma ocorrer como se pode ver na imagem abaixo:

Isto significa que uma imagem 640x640, com 409.600 pixels, não é uma matriz com 640 linhas e 640 colunas, mas sim um vetor com 409.600 elementos. De início, pode parecer uma má notícia já que não podemos simplesmente acessar o último pixel pela posição $x=639$, $y=639$, e nem o primeiro pela posição $x=0$, $y=0$. Mas calma lá! Existe um modo fácil de permitir que você acesse os elementos com facilidade, ainda que estejam em uma estrutura unidimensional, como o nosso vetor. Para qualquer valor de $x$, $y$ na imagem, a localização $p$ do pixel estará em $p = x + y*largura$, sendo $largura$ a largura da imagem. A imagem abaixo esclarece esse procedimento:

Alguns modos de cores:

* 1 (1-bit pixels, black and white, stored with one pixel per byte)
* L (8-bit pixels, black and white)
* P (8-bit pixels, mapped to any other mode using a color palette)
* RGB (3x8-bit pixels, true color)
* RGBA (4x8-bit pixels, true color with transparency mask)
* CMYK (4x8-bit pixels, color separation)
* YCbCr (3x8-bit pixels, color video format)
* Note that this refers to the JPEG, and not the ITU-R BT.2020, standard
* LAB (3x8-bit pixels, the L*a*b color space)
* HSV (3x8-bit pixels, Hue, Saturation, Value color space)
* I (32-bit signed integer pixels)
* F (32-bit floating point pixels)

Pode-se ver que mesmo limitada à tons de cinza, a imagem ainda manteve um aspecto de tons bem diversos. Incrível, não é? Que tal binarizarmos essa imagem? O ato de binarizar uma imagem significa limitar os pixels a apenas duas cores. Uma técnica bastante conhecida é a _threshold global_, isto é, tons acima de um determinado threshold ficarão brancos e tons abaixo de um threshold ficarão pretos. Isto porque o RGB e seus variantes é um espaço de cor aditivo. O menor valor possível de se representar é preto e o maior é branco. O código abaixo converte nossa imagem para uma imagem binarizada com <a href="https://en.wikipedia.org/wiki/Dither" target="_blank">dither</a> já que cada pixel tem apenas 1 bit para representar sua tonalidade, isto é, 1 (branco) ou 0 (preto).

````python
from PIL import Image
from IPython import display


imagem = Image.open("barba_pentada.png")    # Carregando a imagem
imagem = imagem.convert("1")
imagem.save("barba_pentada_pb.png", "PNG")

from IPython.display import Image
Image(filename='barba_pentada_pb.png') 
````

Mas basta de funções prontas. Vamos desenvolver nosso próprio algoritmo de binarização usando a técnica de threshold global!

````python
from PIL import Image
from scipy.misc import imsave
import numpy


def binarize_image(img_path, target_path, threshold):
    """Binarize uma imagem."""
    image_file = Image.open(img_path)
    image = image_file.convert('L')  # Converter para modo monocromático
    image = numpy.array(image)
    image = binarize_array(image, threshold)
    imsave(target_path, image)


def binarize_array(numpy_array, threshold=200):
    """Binarize um array do numpy."""
    for i in range(len(numpy_array)):
        for j in range(len(numpy_array[0])):
            if numpy_array[i][j] > threshold:
                numpy_array[i][j] = 255
            else:
                numpy_array[i][j] = 0
    return numpy_array
    
# Brinque de ajustar o terceiro argumento da função abaixo. É o treshold!
binarize_image('barba_pentada.png', 'barba_pentada_pb1.png', 127)


from IPython.display import Image
Image(filename='barba_pentada_pb1.png')
````

# Vetor ou matriz?

No início deste material, falamos do caso de termos um vetor unidimensional com as informações de todos os pixels. Felizmente, o Python nos oferece através da biblioteca PIL (Python Image Library) a função _load_ que carrega os pixels da imagem de forma bidimensional, como visualmente nos é representado. Por questões meramente acadêmicas, o código abaixo irá encontrar os pixels no vetor unidimensional através da equação apresentada no início deste material, e comparará com o resultado obtido através de nossa matriz bidimensional obtida através do _load_.


````python
from PIL import Image

imagem = Image.open("torax.jpg").convert("L")

vetor_pixels = list(imagem.getdata())
matriz_pixels = imagem.load()
width, height = imagem.size
acertos = 0
for x in range(width):
    for y in range(height):
        p = x+(y*width)
        if vetor_pixels[p] != matriz_pixels[x,y]:
            print "Os pixels comparados possuem valores diferentes."
        else:
            acertos = acertos + 1

if acertos == width*height:
    print 'deu tudo certo :-)'

from IPython.display import Image
Image(filename='torax.jpg')
````

**2)**

Pela resolução espacial, é possível concluir que esta imagem tem 1.048.576 pixels. 1024x1024 = 1.048.576 pixels

Dado que trata-se de uma imagem no modo RGB, teremos três canais para cada pixel, isto é, 24 bits (3x8) ou 3 bytes. Esse valor também é conhecido como _color depth_, comumente descrito como _bpp_, bits ou bytes per pixel.

Multiplicando nosso número de pixels pela quantidade de bytes por pixel temos: 1.048.576 x 3 = 3.145.728 bytes.

Para encontrar o valor de kilobytes é só dividir por 1024. E para encontrar o valor em megabytes é só dividir por 1024 novamente. 3.145.728 / 1024 = 3072K / 1024 = 3M

Portanto, nossa imagem ocupará cerca de 3 megabytes de memória em disco. Mas por que cerca? Pois, como sabemos, as imagens costumam ser codificadas em formatos como JPEG ou PNG que costumam comprimir as imagens. Desse modo, sejam algoritmos de compressão com perda ou não, o tamanho costuma variar. No caso de nossa imagem ser RGBA, teríamos mais uma camada (+ 8 bits por pixel). Esse A, também conhecido como Alpha, é uma máscara de transparência que em um extremo de seu valor deixa a imagem invisível (opaca) ou completamente transparente.

**3)**

Repetindo os cálculos da questão anterior, de forma mais direta, temos:

640x640 = 409.600 pixels

Por ser RGBA, 4x8 = 32 bpp ou 4 bytes per pixel.

409.600 pixels x 4 bytes = 1.638.400 bytes

1.638.400 / 1024 = 1600K / 1024 = 1,6M

````python
from PIL import Image

imagem = Image.open("barba_pentada.png")
matriz_de_pixels = imagem.load()

width, height = imagem.size

for x in range(width):
    for y in range(height):
        r, g, b, a = matriz_de_pixels[x,y]
        matriz_de_pixels[x,y] = 255, 0, 0, a

imagem.save("barba_pentada_r.png")


from IPython.display import Image
Image(filename="barba_pentada_r.png")
````
